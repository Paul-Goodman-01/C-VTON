```python
# Get the current working directory and the data directory
base_path = os.getcwd()
viton_path = base_path+'\\data\\viton'
data_path = viton_path+'\\data'
print(f"Base path: {base_path} Viton_path: {viton_path} Data path: {data_path}")
```

```python
# Get list of directories in the data directory
sub_dirs = [x[0] for x in os.walk(data_path)]
sub_dirs = [os.path.basename(os.path.normpath(item)) for item in sub_dirs]
sub_dirs
```

```python
# Create a dictionary to store data frames for each subdirectory
dataframes = {}
idx = 0 

# Traverse through each subdirectory
for dirpath, _, filenames in os.walk(data_path):
    # Create a data frame for files in the current subdirectory
    files_data = []
    for filename in filenames:
        # Process each file, e.g., read data and create a DataFrame
        if sub_dirs[idx] == 'pose':
            filename = os.path.splitext(filename)[0]
            filename = filename.replace('_keypoints', '')
            files_data.append({'file_name': filename})
        else:
            filename = os.path.splitext(filename)[0]
            files_data.append({'file_name': filename})

    # Create a data frame from the collected data
    if files_data:
        subdir_df = pd.DataFrame(files_data)
        # Use the subdirectory path as the key to store the dataframe
    else:
        subdir_df = pd.DataFrame()
        subdir_df['file_name'] = ' '

    dataframes[sub_dirs[idx]] = subdir_df

    idx += 1
```

```python
samples={}
for f in sub_dirs:
    samples[f] = dataframes[f].shape[0]    

samples
```

```python
sub_dirs_by_samples = {}

for k, v in samples.items():
    if v not in sub_dirs_by_samples:
        sub_dirs_by_samples[v] = [k]
    else:
        sub_dirs_by_samples[v].append(k)

for v, k in sub_dirs_by_samples.items():
    print(f"Sub-dirs for sample count {v}: {k}")
```

```python
# Handle the '2032' samples
group_dfs = []
group_df_keys = sub_dirs_by_samples.get(2032)

for f in group_df_keys:
    group_dfs.append(dataframes[f]) 

# Dictionary to store occurrence information
string_occurrences = {}

# Iterate through each dataframe and its columns to track string occurrences
for i, df in enumerate(group_dfs):
    for index, value in df['file_name'].items():
        if isinstance(value, str):
            if value not in string_occurrences:
                string_occurrences[value] = [group_df_keys[i]]
            else:
                string_occurrences[value].append(group_df_keys[i])

# Print the report
#for string, occurrences in string_occurrences.items():
#    print(f"String '{string}' found in: ".join(occurrences))

vton_result = pd.DataFrame(list(string_occurrences.items()), columns=['file_name', 'sub_dirs'])
```
